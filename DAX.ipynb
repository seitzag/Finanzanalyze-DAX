# Meine DAX Analyze:

```python
url = 'http://real-chart.finance.yahoo.com/table.csv?s=%5EGDAXI&a=00&b=01&c=2014&d=05&e=19&f=2015&g=d&ignore=.csv'
```

```python
import urllib2
```

```python
ge_CSV = urllib2.urlopen(url)
```

```python
import pandas
```

```python
ge = pandas.read_csv(ge_CSV, index_col=0, parse_dates=True)
```

```python
ge.plot(y='Adj Close')
```
